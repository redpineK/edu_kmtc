# 함수
```
def 함수이름(파라매터):
  함수 내용...
```
- 함수 내용에 해당하는 부분(코드 블럭)을 표시하기 위해 들여쓰기를 합니다
- 우리가 사용해왔던 print(), input() 등도 모두 함수
- 반복되는 코드에 이름을 붙여서 다시 사용할 수 있게 합니다.
- 필요할 때 함수의 이름을 불러서 사용할 수 있습니다.
- 이와 유사하게 입력 값이 함수에 들어가면 출력 값이 나옵니다.
- 함수에게 전달하는 값을 인자 혹은 매개변수라고 합니다.
- 굳이 따지자면 다른 용어이지만 보통 혼용해서 사용합니다.
- 함수에게 받아오는 값을 리턴 값, 반환 값, 결과 값이라고 합니다.
- 인자와 결과 값은 있을 수도 있고 없을 수도 있습니다.


````
print("name : {}, age: {}". format("tom", 30) )  
````

## 함수의 반환 값
- 반환하는 값을 원하면 return을 사용한다.
```
def 함수이름(파라미터):
    함수 내용...
    
    return some
```

````
a = int(input())
div = int( a / 3 ) 
print( "입력 값 : ", a, "몫 : ", div )
````

````
def div_fun(a):
    div = int( a / 3 )
    return div

print(div_fun(10))
print(div_fun(15))
````

- 문제2. div_fun함수를 정의하는데 파라메터 2개(a, b) 받아 a / b  의 몫을 구하는 함수를 작성하시오.

````
def div_fun(a, b):
    div = int(a/b)
    return div

print( div_fun(10, 2) )
print( div_fun(6, 3) )
````

````
def div_fun(a, b):
    div = int(a/b)
    mod = a - ( div * b)
    return div, mod

print( div_fun(10, 2) )
print( div_fun(6, 3) )
````

````
def mod_fun(a, b):
    div = int(a/b)
    mod = a - ( div * b )
    return mod
````

## 짝수여부 판단

````
def is_even(n):
    if mod_fun(n,2) == 0:
        return True
    else:
        return False
````

````
print(is_even(5))
````

````
def is_even(n):
    re_even = (mod_fun(n,2) == 0)
    return re_even

print(is_even(3))
````

### 종합문제
- for문을 활용하여 1~50까지 더한 값을 a라는 변수에 저장한 후, 
a / 3으로 나눈 몫과 나머지를 출력하는 함수를 
작성하시오

- 출력
- 더한 값: 00, 몫: 00, 나머지: 00 

- 원하는 값을 입력으로 받아 더한 값 a


````
def sum_fun():
    a = 0
    for i in range(51):
        a = a + i
        
    div = int(a/3)
    mod = a - ( div * 3)
    
    print("더한 값 : {}, 몫 : {}, 나머지 : {}".format(a, div, mod))

          
sum_fun()
````

````
def sum_fun():
    a = 0
    for i in range(51):
        a = a + i
        
    div = int(a/3)
    mod = a - ( div * 3)
    
    return a, div, mod

print(sum_fun())

print("더한 값 : {}, 몫 : {}, 나머지 : {}".format(sum_fun()[0], sum_fun()[1], sum_fun()[2]))  
````

# 함수 추가 설명

- 쥬스 전문점
- 여러가지 과일을 갈아서 과일쥬스를 만들어 준다
- 단일 과일, 딸기 바나나, 딸기오렌지 등 믹서기 안에 넣는 종류에 따라 결과물이 달라진다.

- 쥬스라는 출력을 얻기 위해 과일을 주재료로 한 갖가지 재료를 믹서기에 입력으로 넣는다
- 믹서기 : 함수
- 과일 : 입력
- 과일쥬스 : 출력

- 즉 어떤 과일(입력)을 받아 '갈기'라는 작업(함수안에서 정의해논)을 수행하고 그 결과(과일쥬스)를 반환해 주었다.
- 어떤 입력에 대한 작업의 결과를 출력해 주는 것

- return은 직원용 과 손님용
- return이 없다면 직원들이 목마를 때 스스로 갈아서 마시는 자급자족 이라면
- return이 있을시에는 손님에게 전달하여 손님이 마시거나 다른사람에게 전달하거나 마음대로 다룰 수 있는 전달 된 쥬스라는 것이다.
