# 연습문제1
- name 이라는 변수에 나의 이름을 할당
- age라는 변수에 나의 나이를 할당
- 출력 : 나의 이름은 OOO 입니다. 나이는 OO살 입니다.

````
name = "김정수"
age = "33"
print("나의 이름은" , name, "입니다. 나이는" , age, "살 입니다." )
````

````
print("나의 이름은 " + name + " 입니다. 나이는 " + age + " 살 입니다.")
````
