## (2) mod함수, is_even함수를 통해 짝,홀수의 합을 계산해 보겠습니다

````
def mod_fun(a,b):
    div = int(a/b)
    mod = a - ( div * b )
    return mod

print(mod_fun(4,2))
````

````
def is_even(n):
    re_even = ( mod_fun(n,2) == 0 )
    return re_even

print(is_even(4))
````

````
num = int(input())
if is_even(num):
    print("짝수")
else:
    print("홀수")
````

## 특정한 수의 배수 더하기
- 0~10까지의 2의 배수 더하기
- 2 + 4 + 6 + 8 + 10 = 30

````
def is_multiple(a,b):
    div = int(a/b)
    mod = a - ( div * b )
    return mod == 0

print( is_multiple(4,2) )
print( is_multiple(4,3) )
print( is_multiple(6,3) )
print( is_multiple(6,4) )
````

- 0~10 까지의 2의 배수 더하기

````
sum = 0
for i in range(11):
    if is_multiple(i,2):
        sum=sum+i

print(sum)
````

## 논리연산자
- 연산자 앞과 뒤의 값을 비교해서 참과 거짓을 반환
- 논리연산자의 앞과 뒤에는 논리 값 불리언 자료형이 오게됩니다.
- and : 앞뒤의 값이 모두 참일 때 참을 반환
- or : 둘중에 하나만 참이어도 참을 반환
- not : 논리연산자의 결과를 뒤집어서 반환

````
print( 6 == 6 )
print( 6 < 3 )
````

````
print( 6 == 6 and 6 > 3 )
print( 6 == 6 and 6 < 3 )
````

````
print( 6 == 6 or 6 > 3 )
print( 6 == 6 or 6 < 3 )
````

````
print( not( 6 == 6 ) )
````

# 구구단 

````
def get_n_star(n):
    stars = ""
    for i in range(n):
          stars = stars + "*"
    return stars

print(get_n_star(1))
print(get_n_star(2))
print(get_n_star(3))
print(get_n_star(4))
````

````
for i in range(1,5):
    print(get_n_star(i))
    
````

# while

- for문처럼 반복횟수가 정해져 있지 않다
- 즉, 무한반복에 빠질 수 있다.

````
i = 0
while i < 5:
    print(i)  
    i = i + 1
````

# 함수
```
def 함수이름(파라미터1, 파라미터2, ...):
    함수 내용
    return result1, result2, ...)
```

````
def function(x):
    a = 3
    b = 5
    y = a * x + b
    return y                     # y 값을 돌려준다

c = function(10)
````

````
c
````

````
def function(x):
    a = 3
    b = 5
    y = a * x + b
    print(y)                     # y 값을 출력한다

c = function(10)
````

````
c
````

### print와 return의 차이
- print는 화면에 결과를 보여주는 역할을 한다. print만으로는 함수의 결과를 프로그램의 다른 부분에서 사용 불가능
- 반면, return문으로 반환된 결과는 프로그램의 다른 부분에서 사용할 수 있다.
