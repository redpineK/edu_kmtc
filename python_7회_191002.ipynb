# 1. Dictionary 복습

````
l = ["cat", "dog", "apple"]
print(l)
print(l[0])
````

````
d = { "cat" : "야옹", 
      "dog" : "멍멍", 
      "apple" : "딜리셔스"
    }
print(d)
print(d["cat"])
````

````
while True:
    word = input()
    if word == "":
        break
    print(d[word])
````

````
a = "TOM"
print(a)
print(a.lower())
````

## 5.파일의 내용을 출력하는 함수 만들기

````
def show_file(file_name):
    f = open(file_name, "r")
    lines = f.readlines()
    f.close()
    
    for i in lines:
        print(i.strip())
````

````
show_file("my_text.txt")
````

- 문제. my_text.txt파일을 읽어서 안에 데이터를 모두 대문자로 변경하여 출력

````
f = open("my_text.txt", "r")
lines = f.readlines()
f.close()
print(lines)
for i in lines:
    line = i.strip()
    line = line.upper() 
    print(line)
````

- 문제. 사용자 입력을 2개를 받아서 name.txt에 저장하는데 다음과 같이 저장되게 하시오
```
cat
dog
```
힌트 : 문자열 연결 연산자 + 와 \n을 write부분에 사용

````
f = open("name.txt", "w")

name = input()
f.write(name+"\n")

name = input()
f.write(name+"\n")

f.close()
````

````
show_file("name.txt")
````

- 문제. 사용자의 입력을 계속 받아서 파일 name.txt에 저장
- 빈줄 입력 받을 시 중지

````
names = []
while True:
    name = input()
    if name == "":
        break
    names.append(name)
print(names)
f = open("name.txt", "w")
for name in names:
    f.write(name + "\n")
f.close()

show_file("name.txt")
````

## 종합문제. 이름과 나이를 입력으로 받아서 people.txt 파일로 저장
- 한줄에 이름과 나이를 저장, 이름과 나이는 콤마( , ) 로 구분
- 빈칸을 입력 받으면 중지
```
tom,33
jerry,22
```

````
f = open("people.txt", "w")
while True:
    name = input()
    if name == "":
        break
    age = input()
    if age == "":
        break
    f.write( name + "," + age + "\n" )
f.close()

show_file("people.txt")
````
