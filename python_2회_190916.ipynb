- 피젯수, 젯수를 입력으로 받아서 몫과 나머지를 구해보도록 하자
- a / b = c
- a : 피젯수 
- b : 젯수

```
a = int( input("피젯수 : ") )
b = int( input("젯수 : ") )
div = int(a/b)
mod = a - ( div * b )

print( "몫은 ", div, "나머지는", mod )
```

- 문제7. 두개의 입력(a, b)을 받아서 곱한 값(c)을 출력한 후, c를 4로 나눈 몫과 나머지를 출력 하시오
- 출력
- 곲한 값 : 00
- 몫 : 00 나머지 : 00

```
a = int(input())
b = int(input())
c = a * b


div = int(c/4)
mod = c - ( div * 4 )
print("곲한 값 : ", c)
print("몫 : ", div, "나머지 : ", mod )
```

- 종합문제
- 4개의 입력(a, b, c)를 입력으로 받아서 a와 b를곱한 값 d를 계산한 후, d / c로 나눈 몫과 나머지를 출력하시오.
-출력 : format 함수를 사용, 몫과 value 같은 값
- 피젯수 : 00, 젯수 : 00, 곱한 값: 00, 몫 : 00, 나머지 : 00, value : 00 

```
a = int(input())
b = int(input())
c = int(input())
d = a * b

div = int(d/c)
mod = d - ( div * c )

print( "피젯수 : {1} , 젯수 : {0}, 곱한 값 : {1}, 몫 : {2}, 나머지: {3}, value : {2}"
      .format(c, d, div, mod) )
```

```
a = int(input())
b = int(input())
c = int(input())
d = a * b

div = int(d/c)
mod = d - ( div * c )

print( "피젯수 : {1} , 젯수 : {0}, 곱한 값 : {1}, 몫 : {2}, 나머지: {3}, value : {2}".format(c, d, div, mod) )
```
