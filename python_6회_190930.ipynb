## (1) 문자열 변환을 활용

- 문제. 문자열 a에 담긴 숫자들의 총 합을 구하시오.
```
a = "10,20,30,40"

힌트
1. 물자열을 split을 통해서 리스트에 저장
2. 저장한 리스트를 for문을 통해서 합계(sum) 처리를 합니다( 문자열 -> 정수로 변환하는 과정이 필요)
```

````
a = "10,20,30,40"
l = a.split(",")
print(l)

l2 = []
for i in l:
    l2.append( int(i))
print(l2)

sum = 0
for i in l2:
    sum = sum + i
print(sum)
````

- 문제1. 이름과 나이를 입력으로 받아서 딕셔너리에 추가하면서 출력. 계속 반복되나, 빈칸이 들어가면 종료

````
d={}
while True:
    name = input("name:")
    if name == "":
        break
    age = int(input("age:"))
    #딕셔너리에 추가하는 줄 예제 ( d["tom"] = 30 )
    d[name] = age
    print(d)
````

# 종합문제
```
d = { "Tom" : 30, "Jerry" : 22, "영희" : 28, "철수" : 37 } 
위와같이 딕셔너리가 정의되어 있다.
소대문자 상관없이 입력으로 받아서 나이를 출력하시오
```

````
d = { "Tom" : 30, "Jerry" : 22, "영희" : 28, "철수" : 37 }

d1 = {}
for name in d.keys():
    lower_name = name.lower()
    age = d[name]
    d1[lower_name] = age
print(d1)

name = input()
name = name.lower()
print(name, d1[name])
````

````
d = { "Tom" : 30, "Jerry" : 22, "영희" : 28, "철수" : 37 }

# d1이라는 빈 딕셔너리 생성
d1 = {}
# for문을 통해 d딕셔너러의 key값을 모두 소문자로 변경하여 저장합니다.
for name in d.keys():
    lower_name = name.lower()
# name에 대한 age값을 구한다
    age = d[name]
# d1에 소문자로 변경한 name값과 age를 추가합니다.
    d1[lower_name] = age

#입력으로 name 변수에 이름을 할당하고
name = input()
#소문자로 변경해줍니다.
name = name.lower()
#입력받은 name으로 d라는 딕셔너리의 나이를 출력합니다
print(name, d1[name])
````
