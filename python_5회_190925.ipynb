## 나누기의 나머지를 mod 함수를 통해서 계산하였으나.
```
- + * /

% : 나눗셈의 나머지를 반환
```



````
def mod_fun(a,b):
    div = int(a/b)
    mod = a - ( div * b )
    return mod

print(mod_fun(5,2))
````

````
print(5 % 2)
````

- 1~10까지의 숫자중 홀수만 리스트 n 에 담기
- % : 나머지를 구하는 산술 연산자

````
n = []

for i in range(1, 11):
    if i % 2 != 0:
        n.append(i)

print(n)
````

### 문제. 1~10까지의 홀수의 전체 합과 평균을 출력하시오.
- n 이라는 list에 홀수를 추가하여 합과 평균을 출력하시오.

````
n = []

for i in range(1, 11):
    if i % 2 != 0:
        n.append(i)
print(n)
sum = 0
for i in n:
    sum = sum + i
mean = sum / len(n)
print(sum)
print(mean)
````

### 종합문제1
- 사용자가 입력한 숫자를 리스트 n에 저장하고, n이라는 리스트에 들어가 있는 숫자 중 3의 배수의 총합과 평균을 구하시오.
- 조건2 : 빈값을 입력하면 종료
```
리스트1을 정의
while 루프  ( while True)
    input으로 입력 받기 
    빈문자열이면 break를 사용하여 빠져나오기 ( if문 == ""이면 break)
    리스트1번째 에 담기 
print("리스트에 담긴 숫자 : ", 리스트1)
```
```
리스트2를 정의
for문에 리스트1을 사용 ( for i in 리스트1 )
    int 변환 ( a = int(i) )
    3의 배수를 체크 ( if a % 3 == 0 )
         리스트2에 append를 통해서 담는다
print("3의 배수: ", 리스트2)
```
```
sum 초기값
for문에 리스트2를 사용해서
    합계구하기
평균을 구하기
출력("합계:", sum, "평균" :", mean )
```
       

````
n1 = []
while True:
    number = input()
    if number == "":
        break
    n1.append(number)
print(n1)
        
n2 = []
for i in n1:
    a = int(i)
    if a % 3 == 0:
        n2.append(a)
print(n2)

sum = 0
for i in n2:
    sum = sum + i
mean = sum / len(n2)
print(sum)
print(mean)
 
````

### 종합문제1
- 사용자가 입력한 숫자를 리스트 n에 저장하고, n이라는 리스트에 들어가 있는 숫자 중 3의 배수의 총합과 평균을 구하시오.
- 조건1 : 빈값을 입력하면 종료
- 조건2 : 리스트에 입력으로 받는 숫자가 8개이상이면 종료

````
n1 = []
while True:
    number = input()
    if number == "":
        break
    n1.append(number)
    if len(n1) >= 8:
        break
print(n1)
        
n2 = []
for i in n1:
    a = int(i)
    if a % 3 == 0:
        n2.append(a)
print(n2)

sum = 0
for i in n2:
    sum = sum + i
mean = sum / len(n2)
print(sum)
print(mean)
 
````
